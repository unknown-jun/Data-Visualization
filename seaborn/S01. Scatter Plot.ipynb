# 산점도(Scatter Plot)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

sns.set(style='whitegrid')
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.relplot(x='bill_length_mm', y='bill_depth_mm', data=penguins);
```

```python
sns.relplot(x='flipper_length_mm', y='body_mass_g', data=penguins);
```

- 두 가지 변수를 입력받아 점이 2차원 배열로 나타내어질 때 다른 변수에 따라 변수를 색칠하면서 한 가지 차원을 추가할 수 있다.
- seaborn에선 이것을 "hue semantic"이라고 하며 점이 색을 통해 의미를 얻기 때문이다.

```python
sns.relplot(x='bill_length_mm', y='bill_depth_mm',
            hue='bill_length_mm', data=penguins);
```

- 구분된 각 클래스 사이의 차이를 강조하기 위해서 style을 이용해 marker를 클래스 별로 설정해줄 수 있다.

```python
sns.relplot(x='flipper_length_mm', y='body_mass_g',
            hue='species', style='species', data=penguins);
```

- size는 클래스 별로 marker의 크기를 정해준다.
- 실수형 변수가 아닌 범주형 값이 들어와도 size를 통해 marker의 크기를 정해줄 수 있다.
- size를 이용해 직접 크기의 범위를 지정해줄 수도 있다.

```python
sns.relplot(x='flipper_length_mm', y='body_mass_g',
            hue='species', size='species', data=penguins);
```

```python
sns.relplot(x='flipper_length_mm', y='body_mass_g',
            hue='species', style='island', data=penguins);
```

- col 옵션을 선택하면 지정된 속성의 갯수만큼(예에서는 island의 unique 값) 표를 따로 그린다.

```python
sns.relplot(x='flipper_length_mm', y='body_mass_g', col='island',
            hue='species',  size='species', data=penguins);
```

```python
sns.relplot(x='bill_length_mm', y='bill_depth_mm',col='island', 
            hue='species', style='species', data=penguins);
```

hue에 카테고리형 변수가 아니라 연속적인 변수도 넣을 수 있는데, 이 경우 seaborn이 알아서 적절한 값을 넣어서 색깔별로 구분을 해줌

```python
sns.relplot(x='bill_length_mm', y='bill_depth_mm',
           hue='flipper_length_mm', data=penguins);
```

```python
sns.relplot(x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g', data=penguins);
```

```python
sns.relplot(x='bill_length_mm', y='bill_depth_mm', 
            hue='body_mass_g', size='body_mass_g', data=penguins);
```

```python
sns.relplot(x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g',
            size='body_mass_g', sizes=(10, 150), data=penguins);
```

- hue/ style/ size 3가지 한번에 사용하는 것도 가능하지만 이 모든걸 한 자료에 쓰면 어떤 연관성이 있는지 한번에 알아보기 힘듦
