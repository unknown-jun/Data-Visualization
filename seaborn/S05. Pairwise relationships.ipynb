```python
import seaborn as sns

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.pairplot(penguins);
```

```python
sns.pairplot(penguins, hue='species');
```

```python
# pairplot  scatter        
g = sns.PairGrid(penguins)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6);
```
