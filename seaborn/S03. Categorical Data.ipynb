# 범주형 데이터 (Categorical Data)
## 범주형 산점도 (Categorical scatterplots)
- `stipplot( )` (with `kind="strip"` the default)
- `swarmplot( )` (with `kind="swarm"`)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])

penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.catplot(x='species', y='body_mass_g', data=penguins);
```

**jitter**
- jitter옵션을 False로 하면 산점도의 분포가 한 줄로 나온다.

```python
sns.catplot(x='species', y='body_mass_g',
            jitter=False, data=penguins);
```

**kind='swarm'**
- catplot에 kind를 'swarm'으로 지정하면 모든 값들이 겹치지 않게 펴지게 값들을 표현해준다.

```python
sns.catplot(x='species', y='body_mass_g',
            kind='swarm', data=penguins);
```

```python
sns.catplot(x='species', y='body_mass_g',
            hue='sex', kind='swarm', data=penguins);
```

```python
sns.catplot(x='sex', y='body_mass_g',
            hue='species', kind='swarm', data=penguins);
```

```python
sns.catplot(x='sex', y='body_mass_g',
            hue='species', kind='swarm',
            order=['Female','Male'], data=penguins);
```

y축: 범주형  
x축: 연속형  
으로 바꾸기

```python
sns.catplot(x='body_mass_g', y='species',
            hue='island', kind='swarm', data=penguins);
```

```python
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', col='island', kind='swarm',
            data=penguins);
```

```python
sns.catplot(x='species', y='body_mass_g',
            hue='sex', col='island', aspect=.7, 
            kind='swarm', data=penguins);
```

## 범주형 분포도 (Categorical distribution plots):
- `boxplot( )` (with `kind="box"`)
- `boxenplot( )` (with `kind="boxen"`)
- `violinplot( )` (with `kind="violin"`)

### 박스플롯 (Box Plots) 

```python
sns.catplot(x='species', y='body_mass_g',
            kind='box', data=penguins);
```

```python
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', kind='box', data=penguins);
```

**dodge=False**
- dodge=False를 옵션을 써서 hue로 나눠진 막대그래프를 한줄로 세워 비교할 수 있다.

```python
sns.catplot(x='species', y='body_mass_g',
            hue='sex', kind='box', dodge=False, data=penguins);
```

```python
sns.catplot(x='species', y='body_mass_g',
            col='sex', kind='box',
            data=penguins);
```

```python
sns.catplot(x='species', y='body_mass_g',
            row='sex', kind='box', data=penguins);
```

**aspect=(float)**
- aspect에 float나 int인자를 입력하면 boxplot의 크기가 커진다.

```python
sns.catplot(x='body_mass_g', y='species',
            row='sex', kind='box', aspect=4,
            data=penguins);
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.catplot(kind='box', data=iris);
```

**orient='h'**
- orient='h'를 사용하여 x축과 y축을 바꿀 수 있다.

```python
sns.catplot(kind='box', orient='h', data=iris);
```

```python
sns.catplot(x='species', y='sepal_length',
            kind='box', data=iris);
```

```python
sns.catplot(x='petal_length', y='species',
            kind='box', data=iris);
```

### 박슨 플롯 (Boxen Plots)

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.catplot(x='cut', y='price', kind='boxen', data=diamonds);
```

```python
sns.catplot(x='color', y='price', kind='boxen', data=diamonds);
```

```python
sns.catplot(x='color', y='price',
            kind='boxen', data=diamonds.sort_values('color'));
```

```python
sns.catplot(x='clarity', y='price',
            kind='boxen', data=diamonds);
```

### 바이올린 플롯(Violin plots)
- `violinplot`: 커널 밀도 추정과 상자 도표 결합

```python
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', kind='violin', data=penguins);
```

**bw= (float), cut= (float)**

커널 밀도에 대한 세밀한 조정은 bw를 통해 가능함  
cut의 값을 높히면 꼬리는 더 길어짐

```python
sns.catplot(x='species', y='body_mass_g',
            hue='sex', kind='violin',
            bw=.15, cut=10, data=penguins);
```

**split=True**  
split 옵션을 통해 카테고리별 분할이 가능함

```python
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', kind='violin', 
            split=True, data=penguins);
```

**inner**  
plot안에 box가 아니라 원하는 컬럼의 정보를 알고 싶을 때 지정

```python
sns.catplot(x='species', y='body_mass_g',
            hue='sex', kind='violin',
            split=True, inner='stick',
            data=penguins);
```

```python
g = sns.catplot(x='species', y='body_mass_g',
                kind='violin', split=True, inner=None,
                data=penguins)
sns.swarmplot(x='species', y='body_mass_g',
              color='k', size=3,
              data=penguins, ax=g.ax);
```

```python
sns.catplot(kind='violin', data=iris);
```

```python
sns.catplot(kind='violin', orient='h', data=iris);
```

```python
sns.catplot(x='species', y='sepal_length',
            kind='violin', data=iris);
```

```python
sns.catplot(x='petal_length', y='species',
            kind='violin', data=iris);
```

## 범주형 추정치 도표(Categorical estimate plots)
- `barplot()` (with `kind = "bar"`)
- `pointplot()` (with `kind = "point"`)
- `countplot()` (with `kind = "count"`)

### 막대 플롯(Bar plots)

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
sns.catplot(x='origin', y='mpg', ci="sd",
            hue='cylinders', kind='bar', data=mpg);
```

**palette**  
palett=['str']을 써서 색 조합을 바꿀수 있다.

```python
sns.catplot(x='origin', y='horsepower',
            hue='cylinders', kind='bar',
            palette='ch:.20', data=mpg);
```

아래의 그래프에서 검은 선은 오차범위를 나타낸다.

```python
sns.catplot(x='cylinders', y='horsepower',
            kind='bar', palette='ch:.20',
            edgecolor='k', data=mpg);
```

### 포인트 플롯 (Point plots)
- 축의 높이를 사용하여 추정값을 인코딩하여 점 추정값과 신뢰 구간 표시

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.catplot(x='who', y='survived',
            hue='class', kind='point', 
            data=titanic);
```

```python
sns.catplot(x='class', y='survived',
            hue='who', kind='point',
            data=titanic);
```

```python
sns.catplot(x='class', y='survived', hue='who',
            palette={'man':'b','woman':'r', 'child':'g'},
            markers=['^','o','.'], linestyles=['-','--',':'],
            kind='point', data=titanic);
```

```python
sns.catplot(x='embark_town', y='survived',
            hue='who', kind='point', data=titanic);
```

### 카운트 플랏(Count plots)

```python
sns.countplot(y='deck', data=titanic);
```

```python
sns.countplot(y='embarked', data=titanic);
```

```python
sns.countplot(y='class', data=titanic);
```
