# Line Plot

- Plot은 Figure(그림)과 Axes(축)으로 구성
- ```plt.Figure```: 축과 그래픽, 텍스트, 레이블을 표시하는 모든 객체를 포함하는 컨테이너
- ```plt.Axes```: 눈금과 레이블이 있는 테두리 박스로 시각화를 형성하는 플롯 요소를 포함

````python
fig = plt.figure()
ax = plt.axes()
````

````python
plt.plot([1, 2, 3, 4])
plt.ylabel('y-label');
````

````python
plt.plot([1, 2, 3, 4], [1, 6, 14, 16]);
````

- 끝에 세미콜론을 추가하면 [<matplotlib.lines.Line2D at 0x1e19414e100>]와 같은 메세지를 가릴 수 있음
- plot([value]) 등과 같이 parameter를 하나를 쓰게 되면 y의 좌표만을 추가한다.
    - 리스트의 값들이 y값들이라고 가정하고, x값[0, 1, 2, 3]을 자동으로 만들어낸다.
    - x 축은 value의 갯수로 표현이 된다(만약 x축 값이 없다면)
- x 축을 쓸경우 y 축 갯수와 일치해야 한다.

````python
t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--',
         t, t**2,'bs',
         t,t**3,'g^');
# 빨간 대쉬:   x축 = t, y축= t
# 파란 사각형: x축 = t, y축= t**2
# 녹색 삼각형: x축 = t, y축= t**3
````

````python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
# 다른 자료형이 나왔기 때문에 자동으로 matplot에서 다른 색상을 지정해줌
````

## 스타일 지정하기
x, y값 인자에 대해 선의 색상과 형태를 지정하는 포맷 문자열을 세번째 인자에 입력

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20]);
````

* 기본 포맷 분자열은 'b-'인데 파란색('blue')의 선('-')을 의미함
* 'ro'는 빨간색('red')의 원형('o')마커를 의미함
* 또한, axis( )를 이용해서 축의 범위 [xmin, xmax, ymin, ymax]를 지정함

````python
plt.plot(np.random.randn(50).cumsum(), linestyle='-')
plt.plot(np.random.randn(40).cumsum(), linestyle='--')
plt.plot(np.random.randn(30).cumsum(), linestyle='-.')
plt.plot(np.random.randn(20).cumsum(), linestyle=':');
# -, --, -., : 과 같이 부호로도 스타일을 표기할 수 있다.
````
