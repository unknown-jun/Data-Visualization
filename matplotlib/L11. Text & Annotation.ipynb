# Text & Annotation
## Text

![image](https://user-images.githubusercontent.com/70187490/125767065-ef1674f1-5a8f-4623-8378-ab308f1c6088.png)

```python
import matplotlib.pyplot as plt
import numpy as np

a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.figure(figsize=(11,7))
plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(1.0, 0.35, '2.0 * np.random.randn(10000) + 1.0')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.2, 'np.random.standard_normal(10000)')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(5.0, 0.08, 'np.random.rand(50000)-10.0');
```

## Text Style

![image](https://user-images.githubusercontent.com/70187490/125768338-e237f1d1-82f9-4644-9df9-66f7312ab7dd.png)

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

font1 = {'family':'serif',
         'color':'darkred',
         'weight':'normal',
         'size':16}

font2 = {'family':'Times New Roman',
         'color':'blue',
         'weight':'bold',
         'size':12,
         'alpha': 0.7}

font3 = {'family':'Arial',
         'color':'forestgreen',
         'style':'italic',
         'size': 14}

plt.figure(figsize=(11,7))
plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(1.0, 0.35, 'np.random.randn()', fontdict=font1)

plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.2, 'np.random.standard_normal()', fontdict=font2)

plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(5.0, 0.08, 'np.random.rand()', fontdict=font3);
```

## rotating text

![image](https://user-images.githubusercontent.com/70187490/125770173-3a16ac2d-aa95-4a7c-a6c6-3701be459c44.png)

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

font1 = {'family':'serif',
         'color':'darkred',
         'weight':'normal',
         'size':16}

font2 = {'family':'Times New Roman',
         'color':'blue',
         'weight':'bold',
         'size':12,
         'alpha': 0.7}

font3 = {'family':'Arial',
         'color':'forestgreen',
         'style':'italic',
         'size': 14}

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(-3.0, 0.15, 'np.random.rand()', fontdict=font1, rotation=85)

plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.0, 'np.random.standard_normal()', fontdict=font2, rotation=-60)

plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(-10.0, 0.08, 'np.random.rand()', fontdict=font3);
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(3, 6, ". transData(3, 6)", transform = ax.transData)
#     text  
ax.text(0.2, 0.4, ". transAxes(0.2, 0.4)", transform=ax.transAxes)
#     text  
ax.text(0.2, 0.2, ". transFigure(0.2, 0.2)", transform=fig.transFigure);
#       text  
```

```python
# TransAxes   0.2 / 0.4    
# xlim, ylim     .
ax.set_xlim(-6, 10)
ax.set_ylim(-6, 10)
fig
```

```python
x = np.arange(1, 40)
y = x * 1.1
plt.scatter(x, y, marker='.')
plt.axis('equal')
plt.annotate('interestion point', xy=(4, 5), xytext=(20, 10),
             arrowprops=dict(shrink=0.05));
```

```python
x1 = np.random.normal(30, 3, 100)
x2 = np.random.normal(20, 3, 100)
x3 = np.random.normal(10, 3, 100)

plt.figure(figsize=(11, 9))
plt.plot(x1, label='p1')
plt.plot(x2, label='p2')
plt.plot(x3, label='p3')

plt.legend(bbox_to_anchor=(0., 1.0, 1.0, 0.102), loc=0, ncol=3, mode='extend', borderaxespad=0.0)
plt.annotate("important value",(50, 20), xytext=(5, 40), arrowprops=dict(arrowstyle='->'))
plt.annotate("incorrect value",(40, 30), xytext=(50, 40), arrowprops=dict(arrowstyle='->'));
```
