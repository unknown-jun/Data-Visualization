````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

# Line Plot

- Plot은 Figure(그림)과 Axes(축)으로 구성
- ```plt.Figure```: 축과 그래픽, 텍스트, 레이블을 표시하는 모든 객체를 포함하는 컨테이너
- ```plt.Axes```: 눈금과 레이블이 있는 테두리 박스로 시각화를 형성하는 플롯 요소를 포함

````python
fig = plt.figure()
ax = plt.axes()
````

````python
fig = plt.figure( )
plt.plot([0, 0.2, 0.4, 0.6, 0.8, 1] * 5);
# 끝에 세미콜론을 추가하면 [<matplotlib.lines.Line2D at 0x1e19414e100>]와 같은 메세지를 가릴 수 있음
# plot([value]) 등과 같이 parameter를 하나를 쓰게 되면 y의 좌표만을 추가한다.
# x 축는 value의 갯수로 표현이 된다.
````

````python
x = np.arange(0, 10, 0.01)

fig = plt.figure()
plt.plot(x, np.sin(x))
````

````python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
# 다른 자료형이 나왔기 때문에 자동으로 matplot에서 다른 색상을 지정해줌
````

````python
plt.plot(x, np.arange(90, 100, 0.01))
# plot(x, y)를 지정해줄 때 x와 y의 (차원)크기를 같게 해야 한다.
plt.plot(np.arange(10, 1, -1), [4, 9, 1, 2, -5, 6, 7, 12, -9]);
````

````python
plt.plot( np.random.randn(50).cumsum() );
# random값에 대한 누적합을 plot으로 표현
````

# Line Style

|문자|스타일|
|:-:|:-:|
|```'-'```|```'solid'```|
|```'--'```|```'dashed'```|
|```'-.'```|```'dashdot'```|
|```':'```|```'dotted'```|

````python
plt.plot(np.random.randn(50).cumsum(), linestyle = 'solid')
plt.plot(np.random.randn(40).cumsum(), linestyle = 'dashed')
plt.plot(np.random.randn(30).cumsum(), linestyle = 'dashdot')
plt.plot(np.random.randn(20).cumsum(), linestyle = 'dotted');
````

````python
plt.plot(np.random.randn(50).cumsum(), linestyle='-')
plt.plot(np.random.randn(40).cumsum(), linestyle='--')
plt.plot(np.random.randn(30).cumsum(), linestyle='-.')
plt.plot(np.random.randn(20).cumsum(), linestyle=':');
# -, --, -., : 과 같이 부호로도 스타일을 표기할 수 있다.
````

# Color Style
- 색상 문자


|문자|색상|
|:-|:-|
|```'r'```|red|
|```'g'```|green|
|```'b'```|blue|
|```'c'```|cyan|
|```'m'```|magenta|
|```'y'```|yellow|
|```'k'```|black|
|```'w'```|white|

````python
plt.figure(figsize=(10, 8))
plt.plot(np.random.randn(50).cumsum(), 
         color='g', 
         linestyle = '-')

plt.plot(np.random.randn(40).cumsum(),
         color = '#1243FF', 
         linestyle = '--')

plt.plot(np.random.randn(30).cumsum(),
         color = (0.2, 0.4, 0.6),
         linestyle = '-.')

plt.plot(np.random.randn(20).cumsum(),
         color = 'darkblue', 
         linestyle = ':');
````

````python
plt.plot(np.random.randn(50).cumsum(), color = 'skyblue')
plt.plot(np.random.randn(40).cumsum(), color = 'dodgerblue')
plt.plot(np.random.randn(50).cumsum(), color = 'royalblue')
plt.plot(np.random.randn(50).cumsum(), color = 'k');
````

````python
plt.plot(np.random.randn(50).cumsum(), 'b-')
plt.plot(np.random.randn(50).cumsum(), 'g--')
plt.plot(np.random.randn(50).cumsum(), 'r-.')
plt.plot(np.random.randn(50).cumsum(), 'm:');
# line style과 색상을 한번에 단축해서 쓸수 있다
````
