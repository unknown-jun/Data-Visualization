# Setting x/y axis limit

- xlim( )- x축이 표시되는 범위를 지정하거나 반환
- ylim( )- y축이 표시되는 범위를 지정하거나 반환
- axis( )- x, y축이 표시되는 범위를 지정하거나 반환 ( xlim / ylim을 한번에 지정가능 )

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## xlim( ) / ylim( )

![image](https://user-images.githubusercontent.com/70187490/125636639-67ba0568-8916-42ec-a27c-116260f8c309.png)

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlim(0,5)
plt.ylim(0,20);
```

```python
plt.plot(np.random.randn(50))
plt.xlim(-1, 50)
plt.ylim(-5, 5);
```

## axis( [ ] )
![image](https://user-images.githubusercontent.com/70187490/125637238-424870bf-bfc0-4aee-aaa0-9a4d21c25d69.png)

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.axis([0, 5, 0, 20]);
```

```python
plt.plot(np.random.randn(50))
plt.axis([-1, 50, -5, 5]);
```
