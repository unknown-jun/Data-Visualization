```python
import matplotlib.pyplot as plt
```

# 그래프 영역 채우기
그래프의 영역을 채우는 세가지 함수

* fill_between( )
* fill_betweenx( )
* fill( )

## 좌표를 기준으로 선택 x축 부분을 칠하기  

![image](https://user-images.githubusercontent.com/70187490/125651214-3ee0a55e-8794-4237-b9c1-458ed829fce7.png)

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label', labelpad=7, fontsize=12)
plt.ylabel('Y-Label', labelpad=7, fontsize=12)
plt.fill_between(x[1:3], y[1:3], alpha=0.5);
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label', labelpad=7, fontsize=12)
plt.ylabel('Y-Label', labelpad=7, fontsize=12)
plt.fill_between(x[1:3], y[2:3], alpha=0.5);
```

## 좌표를 기준으로 선택 y축 부분을 칠하기  

![image](https://user-images.githubusercontent.com/70187490/125651770-ed43ada1-fa0d-41ed-b270-43939d4495ee.png)

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label', labelpad=7, fontsize=12)
plt.ylabel('Y-Label', labelpad=7, fontsize=12)
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5);
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label', labelpad=7, fontsize=12)
plt.ylabel('Y-Label', labelpad=7, fontsize=12)
plt.fill_betweenx(x[1:4], y[1:4], alpha=0.5);
```

## 두 그래프 사이의 영역 채우기

![image](https://user-images.githubusercontent.com/70187490/125652272-06ee01fe-4ce4-4b76-aa29-9667d2a95a61.png)

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)

plt.xlabel('X-Axis', labelpad=7, fontsize=12)
plt.ylabel('Y-Axis', labelpad=7, fontsize=12)
plt.fill_between(x[1:3], y1[1:3], y2[1:3], alpha=0.5);
```

## 다각형 영역 채우기 - fill( ) 

![image](https://user-images.githubusercontent.com/70187490/125652705-1f0739d3-a21d-47b1-b5c3-9b7ea3e97506.png)

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.fill([1.9, 1.9, 3.1, 3.1], [1.0, 4.0, 6.0, 3.0], color='lightgray', alpha=0.5);
```
