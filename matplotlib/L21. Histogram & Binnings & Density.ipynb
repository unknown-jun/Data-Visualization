# Histogram
- 히스토그램은 도수분포표를 그래프로 나타낸 것으로, 가로축은 계급, 세로축은 도수(횟수나 개수 등)을 나타냄

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (10,8)
```

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59,
          71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight);
```

```python
plt.hist(weight, label='bins=10')
plt.hist(weight, bins=30, label='bins=30')
plt.legend();
```

## 누적 히스토그램 그리기

![image](https://user-images.githubusercontent.com/70187490/126626952-a26dc811-b225-49a5-81f0-a7b983e799c8.png)

```python
plt.hist(weight, cumulative=True, label='cumulative=True')
plt.hist(weight, cumulative=False, label='cumulative=False')
plt.legend(loc='upper left');
```

## 히스토그램의 종류

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
        80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
weight2 = [52, 67, 84, 66, 58, 78, 71, 57, 76, 62, 51, 79,
        69, 64, 76, 57, 63, 53, 79, 64, 50, 61]

plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.hist((weight, weight2), histtype='bar')
plt.title('histtype - bar', fontdict={'fontsize':18})

plt.subplot(2, 2, 2)
plt.hist((weight, weight2), histtype='barstacked')
plt.title('histtype - barstacked', fontdict={'fontsize':18})

plt.subplot(2, 2, 3)
plt.hist((weight, weight2), histtype='step')
plt.title('histtype - step', fontdict={'fontsize':18})

plt.subplot(2, 2, 4)
plt.hist((weight, weight2), histtype='stepfilled')
plt.title('histtype - stepfilled', fontdict={'fontsize':18});
```

```python
data = np.random.randn(10000)
plt.hist(data);
```

```python
plt.hist(data, bins=50, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='none');
```

```python
x1 = np.random.normal(0, 1, 10000)
x2 = np.random.normal(-5, 3, 10000)
x3 = np.random.normal(5, 2, 10000)
d = dict(histtype='stepfilled', alpha=.5, bins=50)

plt.hist(x1, **d)
plt.hist(x2, **d)
plt.hist(x3, **d);
```

```python
x = np.random.normal(size=50000)
y = x - np.random.normal(size=50000)

plt.hist2d(x, y, bins=50, cmap='OrRd')
plt.colorbar();
```

```python
plt.hexbin(x, y, gridsize=20, cmap='OrRd')
plt.colorbar();
```
