```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')
# 한글 폰트 사용을 위해서 세팅
plt.rc('font', family='Malgun Gothic') 
# plt.rcParams["figure.figsize"] = (6,4) 
# plt.rcParams['line.linewidth'] = 2
# plt.rcParams['lines.color'] = 'r'
# plt.rcParams['axes.grid'] = True

# 기본적으로 초기값을 설정하는 코드
```

# Plot Axis

```python
plt.plot(np.random.randn(50));
```

```python
a = pd.DataFrame({'a': [15, 10, 4, 7],
                  'b': [7, 3, 2, 10],
                  'c': [13, 18, 3, -6]})
plt.plot(a)
plt.xlim(0.5, 2.5)
plt.ylim(0, 15)
plt.legend(a);
```

```python
plt.plot(np.random.randn(50))
plt.xlim(-1, 50)
plt.ylim(-5, 5);
```

```python
plt.plot(np.random.randn(50))
plt.axis([-1, 50, -5, 5]);
# plt.axis를 지정함으로써 xlim/ylim을 쓰지 않고 한번에 축 조정 가능
```

```python
b = pd.DataFrame({'분야': ['파이썬','파이썬','자바','C+','C+','자바'],
                  '가격': [15000, 12500, 23000, 32600, 11520, 27000]}, 
                index=['파이썬 데이터 분석', '파이썬 완벽 가이드',
                      '밑바닥부터 시작하는 자바', '나도 모르는 C+', '게임처럼 배우는 C+', '자바보자 자바'])
b.groupby(b['분야']).sum().plot()
plt.xlim(0,2)
plt.ylim(40000, 60000);
```

```python
plt.plot(np.random.randn(50))
plt.axis('tight');
# 'tight'로 지정하면 여백의 지정을 최소한으로 한다.
```

```python
plt.plot(np.random.randn(50))
plt.axis('equal');
# 여백의 비를 같게 만든다.
```

# Plot Label

```python
plt.plot(np.random.randn(50))
plt.title('Title')
plt.xlabel('x')
plt.ylabel('random.randn');
```

```python
g = pd.DataFrame({'2월 가격':[5000, 3520, 2180],
                  '3월 가격':[4320, 4010, 2030],
                  '4월 가격':[3930, 4250, 2200]}, 
                index =['배추', '무', '마늘'])
g.T.plot()
plt.title("채소 가격 변화")
plt.xlabel("기간")
plt.ylabel("가격");
```

```python
plt.plot(np.random.randn(50), label='A')
plt.plot(np.random.randn(50), label='B')
plt.plot(np.random.randn(50), label='C')

plt.title("Title")
plt.xlabel("x")
plt.ylabel("y")

plt.legend();
```
