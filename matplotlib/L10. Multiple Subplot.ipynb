# Subplot

![image](https://user-images.githubusercontent.com/70187490/125756711-ab668dfe-7e83-4ff1-916b-006c1000fa79.png)


```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0., 5.)
x2 = np.linspace(0., 2.)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title("1st Graph", pad=7, fontsize=12)
plt.ylabel("Damped oscillation", labelpad=7, fontsize=10)

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '-.')
plt.title('2nd Graph', pad=7, fontsize=12)
plt.xlabel("time (s)", labelpad=7, fontsize=12)
plt.ylabel('Undamped', labelpad=7, fontsize=12)

plt.tight_layout();
```

```python
plt.subplot(1,2,1)
plt.plot(x1, y1, '-o')
plt.title('1st Graph', pad=7, fontsize=12)
plt.ylabel("Damped oscillation", labelpad=7, fontsize=12)
plt.xlabel("time (s)", labelpad=7, fontsize=12)

plt.subplot(1,2,2)
plt.plot(x2, y2, '-.')
plt.title('2nd Graph', pad=7, fontsize=12)
plt.xlabel("time (s)", labelpad=7, fontsize=12)
plt.ylabel('Undamped', labelpad=7, fontsize=12)

plt.tight_layout();
```

## Sharing Axis

![image](https://user-images.githubusercontent.com/70187490/125758228-c1dfbb9d-47a5-4abd-b179-bfd1a289f862.png)

```python
x1 = np.linspace(0., 5.)
x2 = np.linspace(0., 2.)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.ylabel("Damped oscillation", labelpad=7, fontsize=10)
plt.xticks(visible=False)  # xticks(visible=False)를 해야만이 첫번째 plot의 x축을 없앨수 있다.

ax2 = plt.subplot(2, 1, 2, sharex=ax1)
plt.plot(x2, y2, '-.')
plt.xlabel("time (s)", labelpad=7, fontsize=12)
plt.ylabel('Undamped', labelpad=7, fontsize=12)

plt.tight_layout();
```

### plot 안에 text 넣기

![image](https://user-images.githubusercontent.com/70187490/125759818-a21a9f43-d3ee-410e-9c43-d04f12611f35.png)

```python
plt.figure(figsize=(10, 10))
for i in range(1, 10):
    plt.subplot(3, 3, i)
    plt.text(0.5, 0.5, str((3, 3, i)), ha='center')
```

```python
fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
# subplots_adjust를 통하여 subplots간의 간격을 조절함

for i in range(1, 10):
    plt.subplot(3, 3, i)
    plt.text(0.5, 0.5, str((3, 3, i)), ha='center')
```

```python
fig, ax = plt.subplots(3, 3, sharex='col', sharey='row')
```

```python
for i in range(3):
    for j in range(3):
        ax[i,j].text(0.5,0.5, str((i,j)), ha='center')
fig
```

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.4)
# subplot에 Grid를 지정하는데 2 x 3의 크기로 생성

plt.subplot(grid[0,0])
plt.subplot(grid[0,1:])
# 1번과 2번을 합쳐 반환
plt.subplot(grid[1,:2])
# 0번과 1번을 합쳐 반환
plt.subplot(grid[1,2]);
```

```python
plt.figure(figsize=(5, 6))

x = range(1, 21)
columns = [ np.random.randn(20) * i for i in range(1, 7)]
i = 0
for c in columns:
    i += 1
    
    plt.subplot(3, 2, i)
    plt.plot(x, c, marker='o', linewidth=1, label=c)
    plt.xlim(-1, 21)
    plt.ylim(c.min()-1, c.max()+1)
```
