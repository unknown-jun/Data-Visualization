# 오차막대(Error Bar)
- 데이터의 편차, 예상 측정 오차를 나타내는데 사용
- 오차 막대는 값의 불확실성을 나타냄

![image](https://user-images.githubusercontent.com/70187490/126616809-d4507fc6-d6b8-4901-82ca-73fa0f54b3ca.png)


```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [2.3, 3.1, 1.7, 2.5]  # 데이터의 편차를 나타내는 리스트

plt.errorbar(x, y, yerr=yerr); 
# yerr의 각 값들은 데이터 포인트의 위/아래 대칭인 오차로 표시
```

## 비대칭 편차 나타내기

![image](https://user-images.githubusercontent.com/70187490/126617138-1fdc1f4d-90f6-4e30-b10c-9b166a0dd88a.png)

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [(2.3, 3.1, 1.7, 2.5), (1.1, 2.5, 0.9, 3.9)]

plt.errorbar(x, y, yerr=yerr)
```

- 데이터 포인트를 기준으로 비대칭인 편차를 표시하려면 (2, N) 형태의 값들을 입력해주면 됨  
- 첫번째 튜플의 값들은 아래 방향 편차, 두번째 튜플의 값들은 위 방향 편차를 나타냄

## 상한/하한 기호 표시하기

![image](https://user-images.githubusercontent.com/70187490/126618702-44972698-2b49-4acc-aebf-8cfdac552720.png)

```python
x = np.arange(1, 5)
y = x**2
yerr = np.linspace(0.1, 0.4, 4)


plt.errorbar(x, y+4, yerr=yerr)
plt.errorbar(x, y+2, yerr=yerr, uplims=True, lolims=True)

upperlimits = [True, False, True, False]
lowerlimits = [False, False, True, True]

plt.errorbar(x, y, yerr=yerr, uplims=upperlimits, lolims=lowerlimits);
```

```python
x = np.linspace(0, 20, 40)
dy = 1
y = np.sin(x) + dy * np.random.randn(40)

plt.figure(figsize=(10, 8))
plt.errorbar(x, y, yerr=dy, fmt='H')
plt.grid(True, axis='y');
```

```python
plt.figure(figsize=(10, 8))
plt.errorbar(x, y, yerr=dy, fmt='s', color='darkblue',
             ecolor='gray', elinewidth=2, capsize=0)
plt.grid(True, axis='y');
```
