# Stem plot
- 바 차트와 유사하지만 폭(width)이 없다. 
- 주로 이산 확률 함수나 자기상관관계(auto-correlation)를 묘사할 때 사용된다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 2*np.pi, 51)
y = np.exp(np.sin(x))

plt.figure(figsize=(10, 6))
plt.stem(x, y, linefmt='gray', bottom=1, use_line_collection=True);
```
