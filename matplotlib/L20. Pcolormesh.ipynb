## Insert 그래프 삽입하기

```python
data = np.random.rand(20,20)
fig, ax = plt.subplots()
plt.pcolor(data, cmap='Blues')

#Insert Graph
axins = ax.inset_axes([0.5, 0.5, 0.45, 0.45])
axins.imshow(data[10:20, 10:20], extent=(10, 20, 10, 20),  cmap='Blues',origin='lower');
```

- ax.inset_axes( )는 현재 Axes에 inset axes를 추가함
- 입력한 리스트 [0.5, 0.5, 0.45, 0.45]는 현재 inset axes가 표시될 영역의 x0, y0, width, height
- imshow를 사용하여 data[10:20, 10:20] 영역을 inset 그래프에 나타나도록 함

```python
data = np.random.rand(20,20)
fig, ax = plt.subplots()
plt.pcolor(data, cmap='Blues')

#Insert Graph
axins = ax.inset_axes([0.5, 0.5, 0.45, 0.45])
axins.imshow(data[10:20, 10:20], extent=(10, 20, 10, 20),  cmap='Blues',origin='lower')

axins.set_xticklabels('')
axins.set_yticklabels('')

for axis in ['top', 'bottom', 'left', 'right']:
    axins.spines[axis].set_linewidth(3)
    axins.spines[axis].set_color('r')
```

- set_xticklabels( )를 통해 inset그래프의 눈금 레이블을 지정함
- axins.spines['top'], ... , axins.spines['right']는 inset그래프의 네 방향의 축
- set_linewidth, set_color를 사용해 선의 두께와 색상을 지정

```python
data = np.random.rand(100,100)
fig, ax = plt.subplots()
plt.pcolor(data, cmap='Blues')

axins = ax.inset_axes([0.5, 0.5, 0.45, 0.45])
axins.imshow(data[10:20, 10:20], extent=(10, 20, 10, 20), cmap='Blues', origin='lower')

axins.set_xticklabels('')
axins.set_yticklabels('')

for axis in ['top', 'bottom', 'left', 'right']:
    axins.spines[axis].set_linewidth(3)
    axins.spines[axis].set_color('r')

indicator = ax.indicate_inset_zoom(axins)
```

- ax.indicate_inset_zoom(axins)를 통해 axins에 표시되는 영역을 ax에 표시함

```python
data = np.random.rand(100,100)
fig, ax = plt.subplots()
plt.pcolor(data, cmap='Blues')

axins = ax.inset_axes([0.5, 0.5, 0.45, 0.45])
axins.imshow(data[10:20, 10:20], extent=(10, 20, 10, 20), cmap='Blues', origin='lower')

axins.set_xticklabels('')
axins.set_yticklabels('')

for axis in ['top', 'bottom', 'left', 'right']:
    axins.spines[axis].set_linewidth(3)
    axins.spines[axis].set_color('r')

indicator = ax.indicate_inset_zoom(axins)

indicator[0].set_linewidth(3)
indicator[0].set_edgecolor('r')
indicator[0].set_color('r')

indicator[1][0].set_linewidth(3)
indicator[1][1].set_linewidth(3)
indicator[1][2].set_linewidth(3)
indicator[1][3].set_linewidth(3)

indicator[1][1].set_color('r')
indicator[1][2].set_color('r');
```

- ax.indicate_inset_zoom(axins)는 그래프 상에 영역을 표시하는 사각형과 직선을 튜플의 형태로 반환함
- indicator[0]는 영역을 표시하는 사각형 패치
- set_linewidth / set_edgecolor / set_color를 이용하여 선의 두께 색상을 지정함
- indicator[1]는 연결선 패치
