# Barplot

![image](https://user-images.githubusercontent.com/70187490/125778516-03203abc-9355-482b-9b40-9f552f8a9062.png)

- 막대그래프( Bar graph )는 범주가 있는 데이터 값을 직사각형의 막대로 표현하는 그래프

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018','2019','2020']
values = [100, 400, 900]

plt.bar(x, years)
plt.xticks(x, years);
````

## Color

![image](https://user-images.githubusercontent.com/70187490/125778742-582b0dad-5f7d-49f8-8ccd-53aa19ecd77e.png)

````python
color = ['r','g','b',['r','g','b']]
years = [2018, 2019, 2020]
values= [100, 400, 900]

plt.figure(figsize=(6, 6))
for i, j in enumerate(color, start=1):
    plt.subplot(2, 2, i)
    plt.bar(years, values, color=j)
    plt.xticks(years)
````

## Width

![image](https://user-images.githubusercontent.com/70187490/125784591-a0e5c270-09ea-4117-af12-97a83ebab968.png)

````python
x = np.arange(3)
years = ['2018','2019','2020']
values = [100, 400, 900]

plt.bar(x, values, width=0.4)
plt.xticks(x, years);
````

````python
x = np.arange(3)
years = ['2018','2019','2020']
values = [100, 400, 900]

plt.bar(x, values, width=0.6)
plt.xticks(x, years);
````

## write value on the bar

````python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot()
plt.barh(x, values)

for p in ax.patches:
    x, y, width, height = p.get_bbox().bounds
    ax.text(width*1.01, y+height/2,width,va='center');
````

## Style

![image](https://user-images.githubusercontent.com/70187490/125785261-3d997cb3-2ea1-4358-a670-b79fbf910a92.png)

````python
x = np.arange(3)
years = ['2018','2019','2020']
values = [100, 400, 900]

plt.bar(x, values, align='edge', edgecolor='lightgray', linewidth=5, tick_label=years);
````

````python
plt.style.use('seaborn-notebook')

height = [ np.random.randn() * i for i in range(1, 6) ]
names = ['A','B','C','D','E']

plt.bar(names, height)
plt.xticks( names, fontweight='bold')
plt.xlabel('group');
````

````python
height = [ np.random.randn() * i for i in range(1, 6) ]
names = ['A','B','C','D','E']
y_pos = np.arange(len(names))

plt.barh(y_pos, height)
plt.yticks(y_pos, names, fontweight='bold')
plt.ylabel('gruop');
````

## Cumsum barplot

````python
bar1 = [12, 28, 1, 8, 22]
bar2 = [28, 7, 16, 4, 10]
bar3 = [25, 3, 23, 25, 17]

bars = np.add(bar1, bar2).tolist()

r = [0, 1, 2, 3, 4]
names = ['A','B','C','D','E']

plt.bar(names, bar1, color='royalblue', edgecolor='white')
plt.bar(names, bar2, bottom=bar1, color='skyblue', edgecolor='white')
plt.bar(names, bar3, bottom=bar2, color='lightblue', edgecolor='white')

plt.xlabel('group', fontweight='bold')
plt.xticks(r, names, fontweight='bold');
````

````python
bar_width = 0.25

bars1 = [14, 17, 9, 8, 7]
bars2 = [14, 7, 12, 4, 10]
bars3 = [21, 4, 24, 13, 17]

r1 = np.arange(len(bars1))
r2 = [ x + bar_width for x in r1 ]
r3 = [ x + bar_width for x in r2 ]
# x축을 0.25 차이로 두어 3개의 그래프가 옆으로 하나씩 나오게 해주게끔함

plt.bar(r1, bars1, color='royalblue', width=bar_width, edgecolor='white', label='r1')
plt.bar(r2, bars2, color='skyblue', width = bar_width, edgecolor='white', label='r2')
plt.bar(r3, bars3, color='lightblue', width=bar_width, edgecolor='white', label='r3')

plt.xlabel('group', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(bars1))], ['A','B','C','D','E'])
plt.legend();
````
