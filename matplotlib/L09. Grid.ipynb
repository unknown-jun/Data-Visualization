# Grid
![image](https://user-images.githubusercontent.com/70187490/125762595-47a5b602-b55d-46f5-acb4-3492fce8d374.png)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.xlim(0, 2)
plt.grid(True);
```

## Axis

![image](https://user-images.githubusercontent.com/70187490/125763085-9c87a42e-66a4-4941-8707-2a146d1cc71d.png)

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y');
```

```python
y = np.arange(0, 2, 0.2)
x = range(1, 11)

plt.plot(x, y, 'bo')
plt.plot(x, y**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, y**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='x');
```

## Change the Style

![image](https://user-images.githubusercontent.com/70187490/125763622-1ae54c0e-ab2c-481b-868f-52e61d426720.png)

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)

plt.grid(True, axis='x', color='red', alpha=0.6, linestyle='--', which='major');
```
