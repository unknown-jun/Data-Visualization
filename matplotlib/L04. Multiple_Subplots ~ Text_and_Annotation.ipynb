```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis');
```

### Padding ( 여백 지정하기 )

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=15)
plt.ylabel('Y-Axis', labelpad=20);
```

### Location

- xlabel() 함수의 loc 파라미터는 X축 레이블의 위치를 지정({‘left’, ‘center’, ‘right’})
- ylabel() 함수의 loc 파라미터는 Y축 레이블의 위치를 지정({‘bottom’, ‘center’, ‘top’})

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis',loc='right')
plt.ylabel('Y-Axis',loc='top');
```

## Title

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, 'k-*')
plt.plot(x, x**3, 'r-.^')

plt.title('Graph Title');
```

### Padding & Loction
- plt.title() 함수의 loc 파라미터를 ‘right’로 설정하면, 타이틀이 그래프의 오른쪽 위에 출력
- {‘left’, ‘center’, ‘right’} 중 선택할 수 있으며 디폴트는 ‘center’입니다.
- pad 파라미터는 타이틀과 그래프와의 간격을 포인트 단위로 설정

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, 'k-*')
plt.plot(x, x**3, 'r-.^')

plt.title("Graph Title", loc='right', pad=20);
```

```python
plt.plot(np.random.randn(50))

plt.title("Title", pad=10)
plt.xlabel("X")
plt.ylabel("Random.randn");
```

## font resize

```python
plt.plot(np.random.randn(50), label='A')
plt.plot(np.random.randn(50), label='B')
plt.plot(np.random.randn(50), label='C')

plt.title("Title", pad=7, fontsize=13)
plt.xlabel('X', labelpad=5, fontsize=13)
plt.ylabel('Random.randn', labelpad=5, fontsize=13);
```
