```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus = False)
```

# Multiple Subplots

```python
ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.5, 0.2, 0.3])
# [0.65, 0.5, 0.2, 0.3]의 위치안에 plot을 반환해줌
```

```python
for i in range(1, 10):
    plt.subplot(3, 3, i)
    plt.text(0.5, 0.3, str((3, 3, i)), ha='center')
```

```python
fig = plt.figure(figsize =(12, 12))
fig.subplots_adjust(hspace = 0.4, wspace=0.4)

strings = ['B','C','D','E','F','G','H','I']
df = pd.DataFrame( {'A':np.random.randint(1, 100, 4)} )

for string in strings:
    df_c = pd.DataFrame({string: np.random.randint(1, 100, 4)})
    df = pd.concat([df, df_c], axis=1)

for i in range(0, 9):
    plt.subplot(3, 3, i+1)
    plt.plot(df.iloc[:,i])
    plt.text(0.5, 0.5, str((3, 3, i)), ha='center')
```

```python
price_df = pd.DataFrame()

fruit = ['수박', '바나나', '사과','배']    

for i in fruit:
    fs = pd.Series(np.random.randint(1000, 10000, 4))
    price_df = pd.concat([price_df, fs], axis=1)
    
price_df.columns = fruit

fig = plt.figure(figsize=(12, 12))
fig.subplots_adjust(hspace = 0.4, wspace=0.4)

for i in range(0, 4):
    plt.subplot(2, 2, i+1)
    plt.plot(price_df.iloc[:,i])
    plt.title(str(fruit[i]))
    plt.ylabel(str(fruit[i] + ' ' + '가격'))
    plt.xlabel(str(fruit[i] + ' ' + '수확률 변화량'))
```

```python
grid =  plt.GridSpec(2, 3, wspace=0.4, hspace=0.4)
# subplot에 Grid를 지정하는데 2x3의 크기로 생성
x = np.random.randint(1, 100, 10)
plt.subplot(grid[0,0])
plt.plot(x)
plt.subplot(grid[0,1:])
plt.plot(np.sin(x))
# 1번과 2번의 크기를 합쳐 반환

plt.subplot(grid[1,2])
plt.plot(np.tan(x))
plt.subplot(grid[1,:2]);
plt.plot(np.sin(x)+np.cos(x));
# 0번과 1번의 크기를 합쳐 반환
```

```python
plt.figure(figsize = (5, 6))

x = range(1, 21)
columns = [np.random.randn(20) * i for i in range(1, 7)]
i = 0
for c in columns:
    i += 1
    
    plt.subplot(3, 2, i)
    plt.plot(x, c, marker = 'o', linewidth=1, label =c)
    plt.xlim(-1, 21)
    plt.ylim(c.min()-2, c.max()+2)
```

# Text and Annotation

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
# subplot의 x축과 y축을 모두 0에서 10까지로 설정
ax.text(7, 9, ". transData(7, 9)", transform=ax.transData)
# 실질적인 데이터의 축으로 따라가게끔 text 위치를 설정

ax.text(0.1, 0.2, ". transAxes(0.1, 0.2)",transform=ax.transAxes)
# 텍스트가 축의 비율로 따라가게끔 text 위치를 설정

ax.text(0.5, 0.5, ". transFigure(0.5, 0.5)", transform=fig.transFigure);
# 텍스트가 전체 그래프의 위치의 비율로 따라가게끔 text 위치를 설정
```

```python
ax.set_xlim(-6, 10)
ax.set_ylim(-6, 10)
fig
```

```python
x = np.arange(1, 40)
y = x * 1.1
plt.scatter(x, y, marker='.')
plt.axis( 'equal' )
plt.annotate('interesting point', xy = (4, 5), xytext=(20, 10), 
            arrowprops=dict(shrink=0.05));
```

```python
x1 = np.random.normal(30, 3, 100)
x2 = np.random.normal(20, 3, 100)
x3 = np.random.normal(10, 3, 100)

plt.figure(figsize = (8, 8))
plt.plot(x1, label='p1')
plt.plot(x2, label='p2')
plt.plot(x3, label='p3')

plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=1, ncol=3, mode='extend', borderaxespad=0.)
plt.annotate("important value", (50, 20), xytext=(5, 40), arrowprops=dict(arrowstyle='->'))
plt.annotate("incorrect value", (40, 30), xytext=(50, 40), arrowprops=dict(arrowstyle='->'));
```
