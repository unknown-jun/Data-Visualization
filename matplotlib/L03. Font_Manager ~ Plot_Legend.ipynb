````python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')

plt.rc('font', family = 'Malgun Gothic')
plt.rc("axes", unicode_minus = False)
````

# Font Manager

````python
set([f.name for f in mpl.font_manager.fontManager.ttflist])
````

````python
font1 = {'family':'Yet R', 'size':24, 'color':'w'}
font2 = {'family':'Times New Roman', 'size':18, 'weight':'bold', 'color':'darkred'}
font3 = {'family':'Verdana', 'size':14, 'weight':'light', 'color':'blue'}

Tidx = pd.period_range('2020-1', end='2020-6', freq='M')
time_data = pd.DataFrame({'수학':[50, 60, 55, 65, 80, 88],
                          '국어':[75, 70, 70, 80, 60, 63],
                          '음악':[33, 40, 94, 67, 55, 80]}, 
                         index = Tidx)
time_data.plot()
plt.title("철수 성적", fontdict = font1)
plt.xlabel("Month", fontdict = font2)
plt.ylabel('score', fontdict = font3)
````

````python
fig, ax = plt.subplots()
ax.plot(np.random.randn(10), '-r', label='A')
ax.plot(np.random.randn(10), '--b', label = 'B')
ax.plot(np.random.randn(10), ':g', label = 'C')
ax.plot(np.random.randn(10), '-.o', label = 'D')
ax.axis('equal')
ax.legend();
````

````python
ax.legend(loc = 'upper center', frameon=False, ncol=2);
# frameon=False: 범례에 있는 테두리를 삭제
# ncol=2: 범례의 layout을 2줄로 만듦
fig
````

````python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad = 5);
# fancybox = 모서리를 둥글게 하는 옵션
# framealpha = 범례의 투명도를 조절한다.
# shadow = 범례에 그림자를 조절한다.
# borderpad = 범례 박스의 여백을 조절한다.
fig
````

````python
plt.figure(figsize=(8, 4))
x = np.linspace(0, 10, 100)
y = np.cos(x[:,np.newaxis] * np.arange(0, 2, 0.2))
lines = plt.plot(x, y)
plt.legend(lines[:3], ['C1','C2','C3'], loc='upper right');
# legend의 모든 요소를 다 뽑아내는 것이 아니라 3개만을 지정하여 'C1','C2','C3'로 이름 지정
````

````python
plt.plot(x, y[:,0], label='c1')
plt.plot(x, y[:,1], label='c2')
plt.plot(x, y[:,2], label='c3')
plt.plot(x, y[:,3:])
plt.legend(framealpha=1, frameon=True, loc='upper right');
# 범레를 불투명하게 함
````

````python
x = np.linspace(0, 20, 100)
I = np.cos(x) - np.cos(x[:, np.newaxis])

plt.imshow(I)
plt.colorbar();
# colorbar를 나타나게 하는 함수
````

````python
plt.imshow(I, cmap='Blues')
plt.colorbar();
````

````python
plt.imshow(I, cmap='RdBu')
plt.colorbar();
````

````python
speckles = ( np.random.random(I.shape) < 0.01)
I[speckles] = np.random.normal(0, 3, np.count_nonzero(speckles))

plt.imshow(I, cmap='RdBu')
plt.colorbar(extend = 'both')
plt.clim(-1, 1);
````

````python
plt.imshow(I, cmap= plt.cm.get_cmap('Blues',5))
# colormap의 색상을 5개로만 반환해줌
plt.colorbar()
plt.clim(-1, 1);
# colorbar의 범위를 -1에서 1로 지정
````
