# horizon & vertical
## horizon


- axhline()    y ( )
    - ,   xmin, xmax  *0 1  * 
    - 0  , 1   .

- hlines()  y, xmin, xmax  ,  (xmin, y)  (xmax, y)   

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4, .5)

plt.plot(x, x+1, 'bo')
plt.plot(x, x**2-4, 'g--')
plt.plot(x,-2*x+3, 'r:')

plt.axhline(4, 0.1, 0.9, color='lightgray', linestyle='--')
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3);
```

## vertical
![image](https://user-images.githubusercontent.com/70187490/125766072-c2cb05c2-2ecc-432e-b6c1-c84a3fafaeae.png)

- axvline()    x ( )
    - ,   ymin, ymax  *0 1  * 
    - 0  , 1   .

- hlines()  x, ymin, ymax  ,  (ymin, x)  (ymax, x)   

```python
x = np.arange(0, 4, .5)

plt.plot(x, x+1, 'bo')
plt.plot(x, x**2-4, 'g--')
plt.plot(x,-2*x+3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--')
plt.vlines(1.8, -3, 2, color='gray', linestyle='solid');
```
