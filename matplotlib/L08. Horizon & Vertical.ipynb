# horizon & vertical
## horizon


- axhline() 함수의 첫번째 인자는 y 값(수평선의 위치)
    - 두, 세번째 인자는 xmin, xmax 값으로서 *0에서 1 사이의 값*을 입력
    - 0은 왼쪽 끝, 1은 오른쪽 끝을 의미합니다.

- hlines() 함수에 y, xmin, xmax를 순서대로 입력하면, 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4, .5)

plt.plot(x, x+1, 'bo')
plt.plot(x, x**2-4, 'g--')
plt.plot(x,-2*x+3, 'r:')

plt.axhline(4, 0.1, 0.9, color='lightgray', linestyle='--')
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3);
```

## vertical
![image](https://user-images.githubusercontent.com/70187490/125766072-c2cb05c2-2ecc-432e-b6c1-c84a3fafaeae.png)

- axvline() 함수의 첫번째 인자는 x 값(수직선의 위치)
    - 두, 세번째 인자는 ymin, ymax 값으로서 *0에서 1 사이의 값*을 입력
    - 0은 아래쪽 끝, 1은 위쪽 끝을 의미합니다.

- hlines() 함수에 x, ymin, ymax를 순서대로 입력하면, 점 (ymin, x)에서 점 (ymax, x)를 따라 수직선을 표시

```python
x = np.arange(0, 4, .5)

plt.plot(x, x+1, 'bo')
plt.plot(x, x**2-4, 'g--')
plt.plot(x,-2*x+3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--')
plt.vlines(1.8, -3, 2, color='gray', linestyle='solid');
```
