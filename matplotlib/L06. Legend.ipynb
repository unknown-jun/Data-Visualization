# Legend


````python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')

plt.xlabel('X-Axis', labelpad=5, fontsize=12)
plt.ylabel('Y-Axis', labelpad=5, fontsize=12)
plt.title('Title', pad=10, fontsize=12)
plt.legend();
````

## N-col

![image](https://user-images.githubusercontent.com/70187490/125646293-eea32203-c520-41b5-a3e1-28ca3ea99e7b.png)

````python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')

plt.title('Title', pad=7, fontsize=12)
plt.xlabel('X-Axis', labelpad=7, fontsize=12)
plt.ylabel('Y-Axis', labelpad=7, fontsize=12)

plt.legend(loc='best', ncol=2);
````

## Setting border

![image](https://user-images.githubusercontent.com/70187490/125646445-173b68a1-9f03-4252-9b1d-8ccc700eaf59.png)

````python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')

plt.title('Title', pad=7, fontsize=12)
plt.xlabel('X-Axis', labelpad=7, fontsize=12)
plt.ylabel('Y-Axis', labelpad=5, fontsize=12)

plt.legend(loc='best', frameon=True, shadow=True);
````

````python
fig, ax = plt.subplots()
# fig ax    
ax.plot(np.random.randn(10), '-r', label='A')
ax.plot(np.random.randn(10), ':g', label='B')
ax.plot(np.random.randn(10), '--b', label='C')
ax.axis('equal')
ax.legend();
````

````python
ax.legend(loc='lower right');
fig
````

````python
ax.legend(loc='upper center', frameon=False, ncol=2);
fig
````

````python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1);
fig
````

````python
plt.figure(figsize=(8, 4))
x = np.linspace(0, 10, 100)
y = np.cos(x[:, np.newaxis] * np.arange(0, 2, 0.2))
lines = plt.plot(x, y)
plt.legend(lines[:3], ['c1','c2','c3'], loc='upper right');
# vales 10  3  'c1','c2','c3'  
````

````python
plt.figure(figsize=(8, 4))
x = np.linspace(0, 10, 100)
y = np.cos(x[:, np.newaxis] * np.arange(0, 2, 0.2))
lines = plt.plot(x, y)
plt.legend(lines[:3], ['c1','c2','c3'], loc='upper right',
           framealpha=1, frameon=True);
# framealpha 1   
````
