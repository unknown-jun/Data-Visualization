* boxplot은 전체 데이터로부터 얻어진 아래의 다섯 가지 요약 수치를 사용해서 그려짐
    - 최소값
    - 제 1사분위 수 (Q1)
    - 제 2사분위 수 또는 중위수 (Q2)
    - 제 3사분위 수 (Q3)
    - 최대값
* 사분위수는 데이터를 4등분한 지점을 말함
    - 제 1사분위 수는 전체 데이터에서 하위 25%에 해당하는 값
    - 제 3사분위 수는 전체 데이터에서 상위 25%에 해당하는 값

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use(['seaborn-notebook'])
plt.rcParams['font.size']=12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fix, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c])
ax.set_ylim(-10, 10)
ax.set_ylabel('Value');
```

- 가운데 박스 모양으로부터 그려지는 중심선을 수염(Whisker)라고 함
- boxplot( ) 함수는 기본적으로 
    - 박스의 위쪽 경계로부터 Q3 + whis * (Q3-Q1)까지
    - 박스의 아래쪽 경계로부터 Q1 - whis * (Q3-Q1)까지
수염을 나타냄

## notch
- ax.boxplot( )의 notch파라미터를 True로 지정하면 중앙값의 95% 신뢰구간을 노치형태로 표시함

![image](https://user-images.githubusercontent.com/70187490/125890351-233fab1c-ad1c-49ac-861e-7c5e8e4afae5.png)

```python
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c], notch=True, whis=2.5)
# whis=2.5로 지정해서, 수염의 길이가 길어지고 수염 범위 밖의 데이터가 표시되지 않음
# IQR을 곱하는 수가 1.5에서 2.5로 변경
ax.set_ylim(-10, 10)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value');
```

## 분위수, 중앙값 등 얻기

```python
fix, ax = plt.subplots()

box = ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5)
ax.set_ylim(-10, 10)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

whiskers = [item.get_ydata() for item in box['whiskers']]
meadians = [item.get_ydata() for item in box['medians']]
fliers = [item.get_ydata() for item in box['fliers']]

print('whiskers: ', whiskers)
print('medians: ', meadians)
print('fliers: ', fliers)
```

## 수평 박스 플롯 그리기

```python
fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c], notch=True, whis=2.5, vert=False)
ax.set_xlim(-10, 10)
ax.set_xlabel('Value')
ax.set_ylabel('Data Type');
```
